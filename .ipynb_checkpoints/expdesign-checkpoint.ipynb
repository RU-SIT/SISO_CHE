```python

def extract_unique_values(path, phase):
    # Initialize sets to store unique values
    maxdops_set = set()
    ds_set = set()
    snr_set = set()
    mod_set = set()
    channel_set = set()
    
    # Regular expression pattern for extracting parameters
    pattern = re.compile(rf"train_{phase}_MAXDopS_(\d+)_DS_([\de.-]+)_SNR_([\d\w-]+)_mod_(\w+)_([\w-]+)\.mat")
    pdb.set_trace()
    for _, _, files in os.walk(path):
        
        for file_name in files:
            match = pattern.match(file_name )
            if match:
                maxdops, ds, snr, mod, channel = match.groups()
                maxdops_set.add(maxdops)
                ds_set.add(ds)
                snr_set.add(snr)
                mod_set.add(mod)
                channel_set.add(channel)
    
    # Convert sets to lists
    maxdops_list = list(maxdops_set)
    ds_list = list(ds_set)
    snr_list = list(snr_set)
    mod_list = list(mod_set)
    channel_list = list(channel_set)
    
    return maxdops_list, ds_list, snr_list, mod_list, channel_list


```

```python
# Example usage
path = "/home/CAMPUS/rghasemi/projects/MyPrivaterepo/data"
phase = 'data'  # replace with the actual phase name(data, labels)
maxdops_list, ds_list, snr_list, mod_list, channel_list = extract_unique_values(path, phase)

print("Unique MAXDopS values:", maxdops_list)
print("Unique DS values:", ds_list)
print("Unique SNR values:", snr_list)
print("Unique mod values:", mod_list)
print("Unique Channel values:", channel_list)
```
